```python
# Supervised Learning (Classification Example)
# Predict Pass/Fail based on study hours

import numpy as np
from sklearn.linear_model import LogisticRegression

# Small dataset (Study Hours vs Pass/Fail)
# 0 = Fail, 1 = Pass
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([0, 0, 0, 1, 1, 1])

# Model
model = LogisticRegression()
model.fit(X, y)

# Prediction
hours = np.array([[4.5]])
prediction = model.predict(hours)

print(model.predict([[3.5]]))
```

```python
# Unsupervised Learning (Clustering)
# Group students based on marks

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Marks in two subjects (Math, Science)
X = np.array([[40, 45],
              [42, 43],
              [90, 85],
              [88, 92],
              [50, 55],
              [85, 87]])

# Model
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

print("Cluster Labels:", kmeans.labels_)

# Plot
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.xlabel("Math Marks")
plt.ylabel("Science Marks")
plt.title("Student Clustering")
plt.show()
```

```python
# Linear Regression (Single Variable)
# Predict marks based on study hours

import numpy as np
from sklearn.linear_model import LinearRegression

# Study hours vs Marks
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([35, 40, 50, 55, 65])

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
predicted_marks = model.predict([[6]])
print(model.predict([[8.5]]))
```

```python
# Multiple Linear Regression (Multi-variable)
# Predict marks based on study hours + attendance

import numpy as np
from sklearn.linear_model import LinearRegression

# [Study Hours, Attendance %]
X = np.array([[1, 60],
              [2, 65],
              [3, 70],
              [4, 80],
              [5, 90]])

y = np.array([35, 40, 50, 60, 75])  # Marks

# Model
model = LinearRegression()
model.fit(X, y)

# Prediction
predicted_marks = model.predict([[6, 95]])
print(model.predict([[9,80]]))
```

```python
# Clustering (Customer Segmentation Example)
# Group customers based on income & spending

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# [Annual Income, Spending Score]
X = np.array([[15, 39],
              [16, 81],
              [17, 6],
              [18, 77],
              [19, 40],
              [20, 76]])

# Model
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(X)

print("Cluster Centers:\n", kmeans.cluster_centers_)

# Plot
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.xlabel("Income")
plt.ylabel("Spending Score")
plt.title("Customer Segmentation")
plt.show()
```
